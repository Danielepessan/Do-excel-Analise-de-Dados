```python
df_principal = pd.read_excel("/home/Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal
```

```python
df_Total_acoes = pd.read_excel("/home/Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_Total_acoes
```

```python
df_Ticker = pd.read_excel("/home/Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_Ticker
```

```python
df_ChatGPT = pd.read_excel("/home/Imersão Python - Tabela de ações.xlsx", sheet_name="ChatGPT")
df_ChatGPT
```

```python
df_principal = df_principal [['Ativo','Data',	'Último (R$)','Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_Final','Var. Dia (%)':'Var_dia_pct' }).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['Var_dia_pct'] / 100
df_principal['Valor Inicial'] = df_principal['Valor_Final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_Total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['Valor_Final'] - df_principal['Valor Inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x >0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtde_teorica'}).copy()
df_principal
```

```python
df_principal = df_principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_ChatGPT, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (anos)':'Ano_de_fundacao'}).copy()
df_principal
```

```python
df_principal['Cat_Ano_de_fundacao'] = df_principal['Ano_de_fundacao'].apply(lambda x: 'Mais de 100' if x >0 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
# Calcular o maior valor da coluna "Variacao_rs"
maior = df_principal['Variacao_rs'].max()

# Calcular o menor valor da coluna "Variacao_rs"
menor = df_principal['Variacao_rs'].min()

# Calcular a média dos valores da coluna "Variacao_rs"
media = df_principal['Variacao_rs'].mean()

# Calcular a média dos valores da coluna "Variacao_rs" para os registros em que o valor da coluna "Resultado" é "Subiu"
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calcular a média dos valores da coluna "Variacao_rs" para os registros em que o valor da coluna "Resultado" é "Desceu"
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

print("Maior\t\t", maior)
print("Menor\t\t", menor)
print("Média\t\t", media)
print("Média de quem subiu\t", media_subiu)
print("Média de quem desceu\t", media_desceu)
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variacao Reais por Resultado')
fig.show()
```
